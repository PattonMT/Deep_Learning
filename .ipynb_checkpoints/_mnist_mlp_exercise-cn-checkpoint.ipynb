# 多层感知器 MNIST
---
在此 notebook 中，我们将训练 MLP 对 [MNIST 手写数字数据库](http://yann.lecun.com/exdb/mnist/)中的图像进行分类。

流程将分为以下几步：
>1. 加载和可视化数据
2. 预处理数据
3. 定义模型
4. 训练模型
5. 保存最好的模型
6. 测试模型

在开始之前，我们需要导入处理数据和使用 PyTorch 所必需的库。

```python
# 导入库
import torch
import numpy as np
```

# 加载和可视化数据
可更改 *batch_size* 来加载数据

```python
from torchvision import datasets
import torchvision.transforms as transforms
from torch.utils.data.sampler import SubsetRandomSampler

# 数据加载时子处理时的数量/是否并行加载数据
num_workers = 0
# 每批量的样本数量
batch_size = 20
# 验证集数量
valid_size = 0.2


# 转化数据为 *torch* 的张量
transform = transforms.ToTensor()

# 设定训练集和测试集
train_data = datasets.MNIST(root='data', train=True, download=True, transform=transform)
test_data = datasets.MNIST(root='data', train=False, download=True, transform=transform)

# 获取训练指针
num_train = len(train_data)
indices = list(range(num_train))
np.random.shuffle(indices)
split = int(np.floor(valid_size * num_train))
train_idx, valid_idx = indices[split:], indices[:split]

# 定义训练和验证样本
train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)

# 准备数据加载
train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size,sampler=train_sampler,num_workers=num_workers)
valid_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size,sampler=valid_sampler,num_workers=num_workers)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=batch_size,num_workers=num_workers)
```

# 可视化一批训练数据
查看数据，确保加载正确

```python
import matplotlib.pyplot as plt
%matplotlib inline
    
# 包含一个批量的训练图片
dataiter = iter(train_loader)
images, labels = dataiter.next()
images = images.numpy()

# 绘制该批量的图片，并将标签呈现
fig = plt.figure(figsize=(25, 4))
for idx in np.arange(20):
    ax = fig.add_subplot(2, 20/2, idx+1, xticks=[], yticks=[])
    ax.imshow(np.squeeze(images[idx]), cmap='gray')
    # print out the correct label for each image
    # .item() gets the value contained in a Tensor
    ax.set_title(str(labels[idx].item()))
```

# 定义网络结构（初始化&前向传播）
将784维向量作为输入，输出为10地张量--图像类别分数

```python
import torch.nn as nn
import torch.nn.functional as F

# 定义网络结构
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # 每个隐藏层地节点数
        hidden_1 = 512
        hidden_2 = 512
        # 784->512
        self.fc1 = nn.Linear(28 * 28, hidden_1)
        # 512->512
        self.fc2 = nn.Linear(hidden_1, hidden_2)
        # 512->10
        self.fc3 = nn.Linear(hidden_2, 10)
        # dropout layer (p=0.2)
        # dropout 防止过拟合
        self.dropout = nn.Dropout(0.2)
        

    def forward(self, x):
        # 将输入图片置平
        x = x.view(-1, 28 * 28) # 变成列形状
        # add hidden layer, with relu activation function
        x = F.relu(self.fc1(x))
        # add dropout layer
        x = self.dropout(x)
        # add hidden layer, with relu activation function
        x = F.relu(self.fc2(x))
        # add dropout layer
        x = self.dropout(x)
        # add output layer
        x = self.fc3(x)
        return x

# initialize the NN
model = Net()
print(model)
```

# 指定[损失函数](https://pytorch.org/docs/stable/nn.html#loss-functions)和[优化器](https://pytorch.org/docs/stable/optim.html)
分类任务：交叉熵损失（pytorch的交叉熵函数向输出层应用了softmax函数）

```python
# specify loss function (categorical cross-entropy)
criterion = nn.CrossEntropyLoss()

# specify optimizer (stochastic gradient descent) and learning rate = 0.01
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
```

# 训练网络
如何用一批数据训练网络：
>1. 清空所有已优化的梯度
2. 前向传播：传入输入，预测输出
3. 计算损失
4. 反向传播：计算损失相对于模型参数的梯度
5. 优化步骤（参数更新）
6. 更新平均训练损失

```python

# 模型训练的周期数
n_epochs = 50
valid_loss_min = np.Inf

for epoch in range(n_epochs):
    # 跟踪训练损失
    train_loss = 0.0
    valid_loss = 0.0
    ###################
    # train the model #
    ###################
    
    # batch loop
    # 训练模式
    model.train() 
    for data, target in train_loader:
        # 清除所有优化变量的梯度
        optimizer.zero_grad()
        # 前向传播
        output = model(data)
        # 计算损失
        loss = criterion(output, target)
        # 反向传播
        loss.backward()
        # 运行优化器，进行权重更新
        optimizer.step()
        # 更新训练损失，loss.item()是平均损失值，乘以size就是累计损失
        train_loss += loss.item()*data.size(0)
        
    ######################    
    # validate the model #
    ######################
    model.eval() # 评估模式
    for data, target in valid_loader:
        # 前向传播
        output = model(data)
        # 计算损失
        loss = criterion(output, target)
        # 更新验证损失
        valid_loss += loss.item()*data.size(0)
             
    # 输出训练数据 
    # 计算每个周期的平均损失
    train_loss = train_loss/len(train_loader.sampler)
    valid_loss = valid_loss/len(valid_loader.sampler)

    print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(
        epoch+1, 
        train_loss,
        valid_loss
        ))
    
    # save model if validation loss has decreased
    if valid_loss <= valid_loss_min:
        print('Validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(
        valid_loss_min,
        valid_loss))
        torch.save(model.state_dict(), 'model.pt')
        valid_loss_min = valid_loss
```
